# Ideas for PML Design

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Finding a Good Enough Decay Function

```python
def compute_decay_profiles(L, N_max, beta=2, m=4):
    X = np.linspace(0, L, N_max)
    sigma_exp = np.exp(beta * (X / L))
    sigma_poly = np.sqrt(3e8) * (X / L) ** m
    sigma_simple = 1 / (L - X) 
    sigma_simple2 = 1 / (L - X) - (1/L)
    return X, sigma_exp, sigma_poly, sigma_simple, sigma_simple2

L = 1.5
N_max = 10
beta = 2
m = 4
X, sigma_exp, sigma_poly, sigma_simple, sigma_s1 = compute_decay_profiles(L, N_max, beta, m)
L = 0.3
X2, sigma_exp2, sigma_poly2, sigma_simple2, sigma_s2 = compute_decay_profiles(L, N_max, beta, m)
```

```python
# Plot and save the first set (PML Thickness = 1.5)
plt.rcParams['font.family'] = 'Verdana'

fig1, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(X[:-1], sigma_s1[:-1], linestyle='-', marker='o', markerfacecolor='none', color='orange', label='Exact Reciprocal 0', linewidth=2, markersize=8)
ax1.plot(X[:-1], sigma_exp[:-1], linestyle=':', marker='v', color='C2', label='Exponential', linewidth=2, markersize=8)
ax1.plot(X[:-1], sigma_poly[:-1], linestyle=(0, (5, 1)), marker='x', color='C0', label='Polynomial', linewidth=2, markersize=8)
ax1.plot(X[:-1], sigma_simple[:-1], linestyle='-', marker='o', color='C3', label='Exact Reciprocal 1', linewidth=3, markersize=8)
ax1.set_xlabel(r'$d_{i,e}$', fontsize=18)
ax1.set_ylabel(r'${\gamma_{i,e}}$', fontsize=18)
ax1.legend(fontsize=12)
ax1.tick_params(axis='both', labelsize=14)
ax1.grid(True)
ax1.set_yscale('log')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("decay_functions.png")
plt.show()

# Plot and save the second set (PML Thickness = 0.3)
fig2, ax2 = plt.subplots(figsize=(8, 6))
ax2.plot(X2[:-1], sigma_s2[:-1], linestyle='-', marker='o', markerfacecolor='none', color='orange', label='Exact Reciprocal 0', linewidth=2, markersize=8)
ax2.plot(X2[:-1], sigma_exp2[:-1], linestyle=':', marker='v', color='C2', label='Exponential', linewidth=2, markersize=8)
ax2.plot(X2[:-1], sigma_poly2[:-1], linestyle=(0, (5, 1)), marker='x', color='C0', label='Polynomial', linewidth=2, markersize=8)
ax2.plot(X2[:-1], sigma_simple2[:-1], linestyle='-', marker='o', color='C3', label='Exact Reciprocal 1', linewidth=3, markersize=8)
ax2.set_xlabel(r'$d_{i,e}$', fontsize=18)
ax2.set_ylabel(r'${\gamma_{i,e}}$', fontsize=18)
ax2.set_yscale('log')
ax2.legend(fontsize=12)
ax2.tick_params(axis='both', labelsize=14)
ax2.grid(True)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("decay_functions_03.png")
plt.show()
```
