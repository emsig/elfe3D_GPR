# Writing Test Input Files for elfe3D_GPR

```python
from elfe3DGPRTestDesignCOAn import elfe3DGPRTestDesign
from elfe3DGPRTestWritingPMLOAn import elfe3DGPRTestWritingPML
import numpy as np
```

## Step 1: Designing the Test

```python
f = 100e6
wave = 3e8/f
print(wave)
```

```python
# Setting Test Parameters
num_points_per_range=1
least_samples_per_wavelength=20
num_layers=2
layer_thicknesses=[1.0, wave/10/3]
eps_r=[1.0, 4.0, 9.0]
sigma=[1e-16, 1e-4, 1e-3]
mu_r=[1.0, 1.0, 1.0, 1.0]
sigma_m=[0.0, 0.0, 0.0, 0.0]
source_type=6
antenna_position=[0.0, 0.0, 0.025]  
w2 = wave/2000
antenna_coordinates=[(-w2, 0.0, 0.02), (w2, 0.0, 0.02)]
current_direction=1
current=1.0
source_moment=1
num_segments=1
ricker_central_f=f
solver_type=2
max_ref_steps=0
max_unknowns=int(5e6)
beta_ref=0.85
accuracy_tol=3e-5
vtk=0
error_est_method=3
ref_strategy=1
num_air_wavelengths=0
num_PML_layers=1
df=10
PML_layer_thickness=wave/10
output_fields_vtk = 1
PML_type = "lin" # "log" or "lin"
PML_theory = 0 # 0: Anisotropic, 1: Differential Operator
PML_decay_type = 1 # 1: Simple Reciprocal, 2: Logarithmic, 3: Exponential, 4: Polynomial
x_e = [-wave/10,1+wave/10]
y_e = [-wave/10,wave/10]
z_e = [-1.0-wave/10/3,wave/10]
experiment_name = "AnAir"
m = 5
s_f = 500
box_x = [-1+0.75, 1+0.375]
box_present = False
bh_f = 1.0 # 2 for lambda/8 box heigh, 1 for lambda/4 box height
anomaly_coordinates = [(0,wave/8),(-wave/20,wave/20),(-0.9,-0.5)]
anomaly_properties = (20, 1e-4, 1.0, 0.0) # (eps_r, sigma, mu_r, sigma_m)
num_rec = 48 #int(np.floor((x_e[1] - 0.4) / ((np.sqrt(3)/2)*(1.5/300.0))))
num_receivers_inline= num_rec
num_receivers_endfire= 0
num_receivers_oblique= 0
```

```python
# Initiating Test Design
test_setup = elfe3DGPRTestDesign(num_points_per_range, least_samples_per_wavelength, num_layers, layer_thicknesses, eps_r, sigma, mu_r, sigma_m, 
                                 source_type, antenna_position, current_direction, current, source_moment, num_segments, 
                                 ricker_central_f, num_receivers_inline, num_receivers_endfire, num_receivers_oblique, 
                                 solver_type, max_ref_steps, max_unknowns, beta_ref, accuracy_tol, vtk, error_est_method, 
                                 ref_strategy, output_fields_vtk, num_air_wavelengths, num_PML_layers, 
                                 PML_layer_thickness, PML_type, PML_theory, PML_decay_type, x_e, y_e, z_e, experiment_name, m, box_x, s_f, box_present,
                                 bh_f, anomaly_coordinates, anomaly_properties
                              )
```

## Step 2: Creating Test Files

```python
# Running Test Writing using the Test Design
test_files = elfe3DGPRTestWritingPML(
    test_setup.frequency_range, 
    test_setup.model_layers, 
    test_setup.regions, 
    test_setup.source_antenna, 
    test_setup.receiver_antenna, 
    test_setup.solver, 
    test_setup.domain, 
    test_setup.io,
    test_setup.PML,
    test_setup.anomaly,
    )
```

```python
test_files.evaluate_all_input_data()
test_files.write_all_input_files()
```
